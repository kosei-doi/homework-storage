# 身体論（心身論）講義 - 課題自動生成

## 使い方
1. 下のセルで課題内容を確認
2. 実行セル（▶️）をクリック → AIが自動的に400字程度の回答を生成
3. 生成された回答をコピーして提出


```python
# 課題内容を表示
with open('assignment.txt', 'r', encoding='utf-8') as f:
    print(f.read())
```

```python
# このセルを実行（▶️をクリック）
import subprocess

command = """@body-theory/assignment.txt の課題に基づいて、@body-theory/input.txt を分析し、400字程度の回答を生成してください。

出力要件：
- 400字程度、高校生レベルの平易な言葉で感想文を作成
- 専門用語を避け、日常的な言葉で説明（例：「整体観」→「全体を見る考え方」、「還元論」→「部分に分けて調べる方法」）
- タイトルなどは付けず、プレーンな文章
- 回答のみを出力
- 文章のつなぎをなめらかに、事実の提示だけで終わらない
- ~が分かった。〜という事を知った。という表現を多用する
"""

try:
    subprocess.run('pbcopy', text=True, input=command, check=True)
    subprocess.run(['osascript', '-e', '''
    tell application "Cursor"
        activate
        delay 0.3
    end tell
    tell application "System Events"
        keystroke "l" using {command down, shift down}
        delay 0.3
        keystroke "v" using command down
        delay 0.3
        keystroke return
    end tell
    '''], check=True)
    print('✅ 実行完了 - AIが回答を生成中...')
except Exception as e:
    print(f'❌ エラー: {e}')
    print('手動: Cmd+Shift+L で新しいチャット → 下記を貼り付け')
    print(f'\n{command}')
```
