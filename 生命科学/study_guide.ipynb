```python
# このセルを実行（▶️をクリック）
import subprocess

command = """生命科学/output/study_guide.htmlを削除し、生命科学/input.md の講義資料を分析し、以下の要件を満たす超わかりやすい学習資料をHTML形式で生命科学/output/study_guide.htmlに生成し、生成した後エラーがないか確認しなさい。

【学習の原理】
以下の2つの学習法を意識して作成してください：

1. 精緻化（詳しく説明する）
   - 単に暗記せず、『なぜ』『どうして』を理解できるように説明
   - プロセスの各ステップの理由を明確に記載
   - 仕組みを誰かに説明できるレベルの理解を促す
   - 図や表を使って視覚的に理解できるようにする

2. メモリーパレス/記憶術
   - 生命科学の構造やプロセスを視覚的に記憶できるようにする
   - 代謝経路は図で描けるようにフローチャートで表現
   - 器官系を『旅する物語』として記憶できるストーリー形式で説明
   - 難しい用語は関連付けのヒントを提供

【出力要件】
- 美しくレイアウトされたHTML出力（CSS付き、印刷可能なデザイン）
- 重要事項は色分けやハイライトで強調
- 図表は可能ならmermaid.jsやSVGで描画
- 各セクションに『なぜそうなるのか？』の説明を必ず含める
- 記憶のヒント（関連付け、視覚的イメージ）を随所に配置
- 構造化して整理
- 数式がある場合はMathJaxを使用
"""

try:
    subprocess.run('pbcopy', text=True, input=command, check=True)
    subprocess.run(['osascript', '-e', '''
    tell application "Cursor"
        activate
        delay 0.3
    end tell
    tell application "System Events"
        keystroke "i" using command down
        delay 0.3
        keystroke "v" using command down
        delay 0.3
        keystroke return
    end tell
    '''], check=True)
    print('✅ Cursor Composer起動 - 学習資料生成中...')
    print('📚 精緻化と記憶術を意識した超わかりやすい学習資料を作成します')
except Exception as e:
    print(f'❌ エラー: {e}')
    print('手動: Cmd+I → 下記を貼り付け')
    print(f'\n{command}')
```
