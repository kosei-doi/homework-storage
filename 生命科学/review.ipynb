```python
# このセルを実行（▶️をクリック）
import subprocess

command = """生命科学/output/review.htmlを削除し、生命科学/worksheet.mdのワークシート問題と生命科学/my_answer.mdの自分の解答を読み込み、以下の要件を満たす添削と解説をHTML形式で生命科学/output/review.htmlに生成してください。

【添削の要件】
1. 各問題について以下を評価：
   - 正解/不正解の判定
   - 部分点の可能性がある場合はその評価
   - 解答の質（完全性、正確性、説明の明確さ）

2. 間違いがある場合：
   - なぜ間違っているのかを明確に説明
   - 正しい答えとその理由を詳しく解説
   - 同じミスを繰り返さないための学習ポイントを提示

3. 正解している場合でも：
   - より良い説明方法や表現方法を提案
   - 関連する知識の補足説明
   - 応用問題への発展のヒント

【解説の要件】
- 精緻化の観点から「なぜその答えになるのか」を詳しく説明
- 記憶術の観点から「どう覚えると良いか」のヒントを提供
- 図や表を使って視覚的に理解できるようにする
- 関連する知識とのつながりを示す
- テストで出題されそうな類似問題のパターンも提示

【出力要件】
- 美しくレイアウトされたHTML出力（CSS付き）
- 問題ごとに色分け（正解=緑、不正解=赤、部分点=黄）
- 各問題に「ポイント」「なぜ間違えたか」「正しい理解」のセクションを設ける
- 総合評価と今後の学習アドバイスを最後に記載
- 数式がある場合はMathJaxを使用
- 印刷可能なデザイン
"""

try:
    subprocess.run('pbcopy', text=True, input=command, check=True)
    subprocess.run(['osascript', '-e', '''
    tell application "Cursor"
        activate
        delay 0.3
    end tell
    tell application "System Events"
        keystroke "i" using command down
        delay 0.3
        keystroke "v" using command down
        delay 0.3
        keystroke return
    end tell
    '''], check=True)
    print('✅ Cursor Composer起動 - ワークシート添削中...')
    print('📝 あなたの解答を詳しく添削・解説します')
except Exception as e:
    print(f'❌ エラー: {e}')
    print('手動: Cmd+I → 下記を貼り付け')
    print(f'\n{command}')
```
